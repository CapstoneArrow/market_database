```python
import pandas as pd
# csv파일 불러오기
df = pd.read_excel('/content/전통시장정보데이터(강원도).xlsx')
```

```python
df
```

```python
address= df['소재지도로명주소']
print(address.head())
```

```python
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
```

```python
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
address_df = pd.DataFrame({'시장명': df['시장명'],'소재지도로명주소':df['소재지도로명주소'],'주소':address,'위도':latitude,'경도':longitude})
```

```python
address_df.to_excel('전통시장정보데이터(강원도-위경도).xlsx')
```
